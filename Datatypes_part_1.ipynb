2.3 Slicing
* Extracts a substring from a string
* Syntax: StartPos: EndPos : SkipCount / Direction
* StarPos -> Included, EndPos -> Excluded
* By Default, slicing always work from left to right unless change by direction

```python
s[0:3]
```

```python
s[3:10]
```

```python
s[:3]
```

```python
s[3:]
```
